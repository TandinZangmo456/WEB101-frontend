```python
# Calculate the mean of the precipitation column
precip_mean = train_df['Precip'].mean()

# Fill the missing values with the mean
train_df['Precip'] = train_df['Precip'].fillna(precip_mean)

train_df.isnull().sum()
```

```python
# assuming 'train_df' is your DataFrame and 'Date' is your column with dates as strings
train_df['Date'] = pd.to_datetime(train_df['Date'])

train_df['Date'] = pd.to_datetime(train_df['Date'], format='%Y-%m-%d')

train_df['Date'] = train_df['Date'].apply(lambda x: x.timestamp())
```

```python
# Convert numeric columns to float
numeric_columns = ['Latitude', 'Longitude', 'Precip']
train_df[numeric_columns] = train_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

print(train_df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Assuming 'station' is a categorical column 
train_df['station_encoded'] = label_encoder.fit_transform(train_df['station'])

# Display the mapping of original labels to encoded labels
print("Mapping of original labels to encoded labels:")
for original_label, encoded_label in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{original_label}: {encoded_label}")

# Display the DataFrame with the new encoded column
print(train_df[['station', 'station_encoded']])
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load your dataset (replace this with the path to your dataset)
train_df = pd.read_csv("/kaggle/input/dsf101-reassessment/train.csv")

# Identify numeric and non-numeric columns
numeric_cols = train_df.select_dtypes(include=['number']).columns
non_numeric_cols = train_df.select_dtypes(exclude=['number']).columns

# Create separate transformers for numeric and non-numeric columns
numeric_transformer = SimpleImputer(strategy='mean')
non_numeric_transformer = SimpleImputer(strategy='most_frequent')

# Create a column transformer to apply transformers to the respective columns
preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', numeric_transformer, numeric_cols),
        ('non_numeric', non_numeric_transformer, non_numeric_cols)
    ]
)

# Create a pipeline with the preprocessor
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fit and transform the data
train_df_imputed = pipeline.fit_transform(train_df)
```
